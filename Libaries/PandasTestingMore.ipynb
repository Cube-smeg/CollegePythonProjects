```python
import pandas as pd 
```

```python
data = {"Name": [], "Age": []}
while True: 
    personName = input("Enter your name")
    personAge = int(input("Enter your age"))

    data["Name"].append(personName)
    data["Age"].append(personAge)
    end = int(input("Enter 1 to exit"))
    if end == 1:
        break
df = pd.DataFrame.from_dict(data)
print(df)
```
