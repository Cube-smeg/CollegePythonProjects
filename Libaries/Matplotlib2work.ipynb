```python
from matplotlib import pyplot as plt
import pandas as pd
file = "Pass-Fail Data.csv"
data = pd.readcsv(file)
mean_data = data.groupby["attendance_pct"]["midterm_score"]["pass"].mean().resetindex()
print(mean_data)

plt.bar(mean_data["attendence_pct"], mean_data["pass"], color="Red")
plt.xlabel("Attendence Rate")
plt.ylabel("Pass or Fail")
plt.title("How passing rates differ by attendence")
plt.xticks(mean_data["attendence_pct"])
plt.show()















```
