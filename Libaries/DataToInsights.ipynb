```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creating datasets for line
hours_studied = [0, 0, 14, 20, 25, 10, 5, 55, 70, 30]
test_scores = [73, 66, 78, 82, 81, 74, 71, 91, 94, 89]
#For bar:
fav_drink = ["Coke","Pepsi", "Monster", "Dr Pepper"]
fav_drink_values = ["21", "17 ", "15","19"]
#For pie
```

```python
#Creating lineplot:
plt.title = "Hours Studied to Test Scores"
plt.xlabel = "Hours Studied"
plt.ylabel = "Test Scores"
plt.plot(hours_studied, test_scores,linestyle="--", color="Green")
plt.show()
```

```python
#Creating Bargraph:
plt.bar(fav_drink_values, fav_drink)
plt.title("Favourite Drinks")
plt.xlabel = "Votes"
plt.ylabel = "Drink"
plt.show()
```
